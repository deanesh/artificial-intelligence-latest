```python
import numpy as np
import random

# Environment settings
states = [0, 1, 2, 3, 4]  # 0=start, 4=goal, 2=hole
actions = ["left", "right"]
rewards = {4: 10, 2: -10}  # rewards for goal and hole
gamma = 0.8  # discount factor
alpha = 0.1  # learning rate
epsilon = 0.2  # exploration rate

# Q-table initialized to zeros
Q = np.zeros((len(states), len(actions)))


# Helper to move in environment
def step(state, action):
    if action == "left":
        next_state = max(0, state - 1)
    else:  # "right"
        next_state = min(len(states) - 1, state + 1)

    reward = rewards.get(next_state, 0)
    done = next_state in rewards
    return next_state, reward, done


# Training loop
episodes = 100

for episode in range(episodes):
    state = 0
    done = False

    while not done:
        # Choose action: explore or exploit
        if random.uniform(0, 1) < epsilon:
            action_index = random.choice([0, 1])  # Explore
        else:
            action_index = np.argmax(Q[state])  # Exploit best known

        action = actions[action_index]
        next_state, reward, done = step(state, action)

        # Q-learning update
        best_next_q = np.max(Q[next_state])
        cur_state_q = Q[state, action_index]
        Q[state, action_index] += alpha * (reward + gamma * best_next_q - cur_state_q)

        state = next_state

# Print learned Q-values
print("Learned Q-table:")
for s in range(len(states)):
    print(f"State {s}: Left={Q[s,0]:.2f}, Right={Q[s,1]:.2f}")
```
