## ✅ **Retail Sales Forecasting with Prophet and ARIMA**

### 📦 Sample Code

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error
from math import sqrt

# Load sample retail sales data
# Replace this with your actual dataset
df = pd.read_csv('retail_sales.csv', parse_dates=['date'])

# Basic preprocessing
df = df[['date', 'sales']]
df = df.rename(columns={'date': 'ds', 'sales': 'y'})
df = df.sort_values('ds')

# Train-test split (e.g., last 3 months as test)
train = df[:-90]
test = df[-90:]

# ------------------ #
# 🔮 Model 1: Prophet #
# ------------------ #
prophet_model = Prophet()
prophet_model.fit(train)

# Forecasting
future = prophet_model.make_future_dataframe(periods=90)
forecast = prophet_model.predict(future)

# Merge predictions with actuals
prophet_pred = forecast[['ds', 'yhat']].set_index('ds').join(df.set_index('ds'))
prophet_rmse = sqrt(mean_squared_error(test['y'], prophet_pred.loc[test.index, 'yhat']))

# Plot
prophet_model.plot(forecast)
plt.title(f'Prophet Forecast | RMSE: {prophet_rmse:.2f}')
plt.show()

# ---------------- #
# 🔄 Model 2: ARIMA #
# ---------------- #
# ARIMA expects a 1D series
arima_model = auto_arima(train['y'], seasonal=True, m=7, trace=True, suppress_warnings=True)

# Forecast
arima_forecast = arima_model.predict(n_periods=90)
arima_rmse = sqrt(mean_squared_error(test['y'], arima_forecast))

# Plot ARIMA results
plt.figure(figsize=(10, 5))
plt.plot(train['ds'], train['y'], label='Train')
plt.plot(test['ds'], test['y'], label='Actual')
plt.plot(test['ds'], arima_forecast, label='ARIMA Forecast')
plt.title(f'ARIMA Forecast | RMSE: {arima_rmse:.2f}')
plt.legend()
plt.show()

# ----------------------------- #
# 📊 Model Comparison Summary   #
# ----------------------------- #
print(f"Prophet RMSE: {prophet_rmse:.2f}")
print(f"ARIMA RMSE: {arima_rmse:.2f}")

if prophet_rmse < arima_rmse:
    print("✅ Prophet performed better.")
else:
    print("✅ ARIMA performed better.")
```
