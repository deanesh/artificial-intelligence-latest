```python
Architecture and working style of Agentic AI.

üß† AI Agents Work:
> Tasks
> Planning
> Execution
Agentic AI systems follow a goal-oriented workflow made of three core phases:

===Example Goal =============================================
- Research top AI tools for designers and create a blog draft
- Summarize this PDF and convert it into a 5-minute video script

===‚úÖ Tasks (What should I do phase) ==================================
Agent plans:
Translates goal into a set of subtasks (breakdown).

===üìã Planning (How do I get it done phase)============================

Agent decides:
Which tools or steps are needed, In what order, Whether it needs to wait for or depend on other results, How to evaluate
progress

üß† Example Planning by an Agent:
Goal: *‚ÄúFind 5 marketing trends and create a LinkedIn post.‚Äù*
Plan:
Search for top marketing trends (2025), Extract the most repeated 5, Write bullet points, Convert to social media tone,
Return formatted post
======================================================================

===‚öôÔ∏è **Execution** (Let me do it phase)==============================

Here, the agent:
Executes each step(one at a time), Calls tools/APIs or code when needed, Refines output if a step fails, May update memory to improve results

üëÄ Example Execution:
Runs a Google Search via API, Extracts data using a scraping tool, Summarizes results using GPT,Posts or emails the final result
```
